# Copy in python
 - ### Shallow
    A shallow copy creates a new list but does not create a copies of nested objects
    
 - ### Deep
    A deep copy creates a new list and new copies of all nested objects

```python
import copy

l4 = [[1, 2], [3, 4]]
print(l4)

shallow_l4 = l4.copy() # shallow copy
deep_l4 = copy.deepcopy(l4) # deep copy

l4[0][0] = 2

print(l4)
print(shallow_l4)
print(deep_l4)
```

Q. Write a program to check whether a given list contains any repeating elements

eg: 
- input in list [1, 2, 3, 4]
    output: False
- input in list [1, 2, 3, 2]
    output: True

```python
# arr = [1, 2, 3, 4]
arr = []

def hasDuplicates(arr):
    for i in range(0, len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] == arr[j]:
                return True
    return False

def main():
    while True:
        try:
            n = int(input("Enter Element:"))
            arr.append(n)
        except ValueError:
            break
    
    print(arr)
    print(hasDuplicates(arr))

if __name__ == '__main__':
    main()
```

Q. Write a program to find the first repeating element in a list

eg: 
- input: [4, 8, 9, 4, 3, 6]
output: 4

```python
arr = []

def getDuplicate(arr):
    for i in range(0, len(arr)):
        print("i:", arr[i])
        for j in range(i + 1, len(arr)):
            print("j:", arr[j])
            if arr[i] == arr[j]:
                return arr[i]
    return "No Repeating Elements"

def main():
    while True:
        try:
            n = int(input("Enter Element:"))
            arr.append(n)
        except ValueError:
            break
    
    print(arr)
    print(getDuplicate(arr))

if __name__ == '__main__':
    main()
```

Q. Write a program to print all the repeating element only once as a new list

```python
arr = []
dupl = []

def getDuplicates(arr):
    for i in range(0, len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] == arr[j]:
                if arr[i] not in dupl:
                    dupl.append(arr[i])
    return dupl

def main():
    while True:
        try:
            n = int(input("Enter Element:"))
            arr.append(n)
        except ValueError:
            break
    
    print(arr)
    print(getDuplicates(arr))

if __name__ == '__main__':
    main()
```

Q. Write a program to count how many times each element occurs

eg:
- input: [1, 2, 1, 3]
    ouput: 1: 2 

```python
arr = [1, 2, 3, 2, 3, 4, 3, 4, 5, 5, 6]
dupl = []

def getDuplicatesFreq(arr):
    for i in range(0, len(arr)):
        count = 1
        for j in range(i + 1, len(arr)):
            if arr[i] == arr[j]:
                count += 1
        dupl_element = (arr[i], count)
        flag = False

        for ele in dupl:
            if dupl_element[0] == ele[0]:
                flag = True

        if not flag:
            dupl.append(dupl_element)

    return dupl


def main():
    # while True:
    #     try:
    #         n = int(input("Enter Element:"))
    #         arr.append(n)
    #     except ValueError:
    #         break
    
    print(arr)
    print(getDuplicatesFreq(arr))

if __name__ == '__main__':
    main()
```

Q. Write a program to print each repeating element along with the list of it's position

eg:
- input: [1, 2, 1, 2, 3]
    output: [(1, [0, 2]), (2, [1, 3]), (3, [4])]

```python
# arr = [1, 2, 3, 2, 3, 4, 3, 4, 5, 5, 6]
arr = [1, 2, 1, 2, 3]
dupl = []

def getDuplicatesLocation(arr):
    for i in range(0, len(arr)):
        loc = [i]
        for j in range(i + 1, len(arr)):
            if arr[i] == arr[j]:
                loc.append(j)
        
        if len(loc) == 1:
            continue
        
        dupl_element = (arr[i], loc)
        flag = False

        for ele in dupl:
            if dupl_element[0] == ele[0]:
                flag = True

        if not flag:
            dupl.append(dupl_element)

    return dupl


def main():
    # while True:
    #     try:
    #         n = int(input("Enter Element:"))
    #         arr.append(n)
    #     except ValueError:
    #         break
    
    print(arr)
    print(getDuplicatesLocation(arr))

if __name__ == '__main__':
    main()
```
